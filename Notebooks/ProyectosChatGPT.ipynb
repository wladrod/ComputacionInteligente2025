### 1. **Asistente Legal Automatizado**

**Objetivo:**  
Crear un agente conversacional que ayude a responder preguntas legales comunes utilizando documentación jurídica como leyes, reglamentos y jurisprudencia.

**Contexto:**  
Los sistemas legales son complejos y muchas personas carecen de acceso claro a información jurídica confiable. Un asistente legal automatizado puede servir como primera guía, sin sustituir el consejo profesional.

**Características esperadas:**
- Capacidad para recuperar documentos relevantes (RAG) ante una consulta legal.
- Respuestas adaptadas al nivel del usuario (experto o ciudadano común).
- Explicación de términos técnicos legales con ejemplos.
- Consideración ética para evitar dar asesoramiento indebido.

### 2. **Sistema de Triage Médico Automatizado**

**Objetivo:**  
Desarrollar un agente capaz de guiar a los usuarios para determinar la gravedad de sus síntomas y sugerirles una posible especialidad médica a consultar.

**Contexto:**  
En contextos donde no hay acceso inmediato a profesionales de salud, los sistemas automatizados pueden ayudar a filtrar y orientar consultas médicas iniciales.

**Características esperadas:**
- Diálogo guiado para recolección de síntomas (simulación de entrevista médica).
- Análisis del riesgo y recomendación de acciones: consulta médica, urgencia, etc.
- Referencia a documentos médicos fiables mediante recuperación.
- Mensajes claros sobre que el sistema **no reemplaza atención médica profesional**.

### 3. **Analizador de Contratos Inteligente**

**Objetivo:**  
Implementar un sistema capaz de leer contratos legales y destacar cláusulas clave, riesgos, y condiciones poco usuales para facilitar su comprensión.

**Contexto:**  
Tanto usuarios comunes como profesionales suelen pasar por alto elementos importantes al revisar contratos extensos y complejos.

**Características esperadas:**
- Extracción automática de secciones del contrato (tokens clave: "renovación automática", "penalización", etc.).
- Resumen claro y resaltado de cláusulas de interés.
- Explicación del impacto legal de ciertos términos.
- Posibilidad de comparar varios contratos.

### 4. **Agente de Productividad Personal**

**Objetivo:**  
Construir un asistente que administre tareas del usuario, organice su día y proporcione recordatorios o sugerencias para mejorar la gestión del tiempo.

**Contexto:**  
Los asistentes personales inteligentes son cada vez más populares entre estudiantes, profesionales y trabajadores remotos.

**Características esperadas:**
- Entrada por texto para registrar tareas y objetivos.
- Resúmenes diarios automáticos y priorización de actividades.
- Capacidad de integrarse con calendarios o correos (simulado o real).
- Recomendaciones basadas en patrones de trabajo del usuario.

---

### 5. **Asesor de Inversión Cripto**

**Objetivo:**  
Crear un agente que recupere y analice información actualizada del ecosistema cripto y brinde orientación sobre tendencias o posibles oportunidades.

**Contexto:**  
El mercado de criptomonedas está sobrecargado de información. Un sistema que sintetice datos y brinde orientación clara puede ser de gran utilidad.

**Características esperadas:**
- Recuperación de datos de mercado o noticias cripto (reales o simuladas).
- Análisis automatizado con lenguaje no técnico.
- Sistema de advertencias éticas y disclaimers legales.
- Opcional: personalización por nivel de riesgo o perfil de usuario.

### 6. **Tutor Personalizado para Cursos Universitarios**

**Objetivo:**  
Desarrollar un tutor virtual que pueda responder preguntas relacionadas con el contenido de una materia universitaria específica, utilizando materiales proporcionados.

**Contexto:**  
Muchos estudiantes no logran resolver sus dudas por falta de tiempo con el docente o materiales poco interactivos.

**Características esperadas:**
- Ingesta de materiales educativos (PDFs, textos, presentaciones).
- Respuestas fundamentadas con citas de las fuentes.
- Prompts diseñados para aclarar conceptos, resolver ejercicios o generar ejemplos.
- Personalización por nivel (novato, intermedio, avanzado).

### 7. **Detector de Sesgos en Noticias**

**Objetivo:**  
Crear un sistema que compare diferentes versiones de una noticia y analice posibles sesgos o diferencias ideológicas entre las fuentes.

**Contexto:**  
La desinformación y el sesgo mediático son preocupaciones clave en el mundo actual. Este sistema busca fomentar el pensamiento crítico.

**Características esperadas:**
- Recuperación de artículos desde diferentes medios sobre el mismo tema.
- Detección de tono, enfoque y omisiones en el contenido.
- Visualización opcional de comparaciones semánticas.
- Prompts que generen explicaciones neutrales sobre los hallazgos.

### 8. **Asistente de Escritura Científica**

**Objetivo:**  
Diseñar un asistente que ayude a los estudiantes e investigadores a estructurar y redactar artículos académicos con estilo formal y coherente.

**Contexto:**  
La escritura científica puede ser una barrera para quienes comienzan en la investigación. Este agente busca acompañar el proceso.

**Características esperadas:**
- Generación de introducciones, objetivos, metodología o conclusiones.
- Recuperación de citas relevantes o papers similares (si es posible).
- Prompts adaptados a lenguaje académico y formato científico.
- Opcional: detección de incoherencias o mejoras de estilo.

---

### 9. **Generador de Cuestionarios Educativos**

**Objetivo:**  
Crear un sistema que genere automáticamente preguntas de opción múltiple, verdadero/falso o abiertas a partir de textos académicos.

**Contexto:**  
Los profesores y creadores de contenido pueden beneficiarse de herramientas que automaticen parte de la evaluación de aprendizajes.

**Características esperadas:**
- Detección de conceptos clave en los textos.
- Generación de preguntas variadas por tipo y dificultad.
- Respuestas correctas automáticas y feedback explicativo.
- Posible personalización por tema o estilo de aprendizaje.

---

### 10. **Simulador de Entrevistas de Trabajo**

**Objetivo:**  
Diseñar un agente que simule entrevistas laborales personalizadas y proporcione retroalimentación para mejorar el desempeño del candidato.

**Contexto:**  
Practicar entrevistas en un entorno controlado es útil para estudiantes y profesionales que se preparan para nuevas oportunidades.

**Características esperadas:**
- Simulación de entrevista por tipo de puesto (desarrollador, analista, diseñador, etc.).
- Evaluación de respuestas por claridad, estructura y contenido.
- Retroalimentación escrita con consejos prácticos.
- Opción de guardar historial de desempeño.
